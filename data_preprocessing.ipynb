```python
import pandas as pd
```

```python
df = pd.read_csv("data/data_without_null_values.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
df.columns
```

```python
cat_cols = ["gender", "diabetic", "smoker"]
label_encoders = {}
```

```python
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

    joblib.dump(le, f"encoders/label_encoder_{col}.pkl")
```

```python
df.to_csv("data/prepared_data.csv", index=False)
```
